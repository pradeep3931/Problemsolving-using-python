```python
#Function to find the height frequency 
def high(k):
    d={}
    for i in k:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    print(d)
    e=list(set(list(d.values())))
    print(max(e))
    for item in  d.items():
        if item[1]==max(e):
            print(item[0])
    
high([1,2,2,2,13,1,4,1,4,5,1,5,5,63])
```

```python
#Function to find lowest frequency in a list 
def low(k):
    d={}
    for i in k:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    print(d)
    e=list(set(list(d.values())))
    lo=[]
    
    for i in d.items():
           if i[1]==min(e):
                lo.append(i[0])
    print(lo[0])
           
           
low([1,2,2,2,13,1,4,1,4,5,1,5,5,63])         
```

```python
#Function to find kth lowest frequency in a list 
def klow(k,l):
    d={}
    for i in k:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    print(d)
    e=list(set(list(d.values())))
    lo=[]
    b=e[l-1]
    print(e)
    for i in d.items():
           if i[1]==b:
                lo.append(i[1])
    print(lo[0])
           
           
klow([1,2,2,2,13,1,4,1,4,5,1,5,5,63],2)         
```
