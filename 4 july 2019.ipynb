```python
#Linear Graph :-
import matplotlib.pyplot as plt
plt.title("Simple linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()#display value
```

```python
#Graph to plot Distence and Time 
Dist=list(range(2,11,2))
time=list(range(1,6))
plt.title("Dist vs Time")
plt.xlabel("time")
plt.ylabel("dist")
plt.plot(time,Dist,'go')# g-color o-points
plt.grid()#For grids
plt.show()


```

```python
#v =a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v,"g")
plt.grid()
plt.show()
 
```

```python
n=list(range(10))
sqn=[i**2 for i in n ]
plt.title("number vs squares")
plt.plot(n,sqn)
plt.grid()
plt.show()
```

```python
# plot income of lowest state

import pandas as pd
data=pd.read_csv("datafiles/income.csv")
years=data.columns[2:]
income=data.values[3,2:]
plt.plot(years,income)
plt.show()
```

```python
#Avarage Income of States
incomeal=[]
for i in data.columns[2:]:
    incomeal.append(data[i].mean())
incomeal
plt.plot(years,incomeal)
plt.show()
```
